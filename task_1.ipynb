#### In a Jupyter Notebook (.ipynb) file, use pandas to load the data from the JSON dataset into a DataFrame. Explore the data. 

```python
# Import dependencies 

import pandas as pd
import json
from pandas.io.json import json_normalize
```

#### Loading the JSON file into a DataFrame

```python
data = pd.read_json('dkdata.json')
data.head()
```

#### The outcome of my first tentative to load the JSON file into a DataFrame returned with a nested column. In order to proceed I need to flatten the whole column and unpack into a new DataFrame. 

```python
# Unpacking nested column into a DataFrame utilizing json_normalize

with open('dkdata.json') as f:
    d = json.load(f)
    
dk_data = json_normalize(d['data'])
dk_data
```
