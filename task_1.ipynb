```python
# Import dependencies 

import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
# Loading json file into a DataFrame

data = pd.read_json('dkdata.json')
data.head()
```

```python
# Checking number of rows 

data.shape
```

```python
# Underlying numpy arrays to see what columns to expect 

data.values
```

```python
# Unpacking nested column into a DataFrame utilizing json_normalize

with open('dkdata.json') as f:
    d = json.load(f)
    
dk_data = json_normalize(d['data'])
dk_data
```

### Basic Analysis 

Now that we were able to properly load all the content of the file into a DataFrame we are going to look for basic informations like, duplicate names, missing values, account numbers etc... 

```python
# Checking for any missing value and summing them up

dk_data.isnull().sum()
```

On the above cell we have learned that there are 115 payers's bank account missing, consequentely 115 payer's bank country missing. Also, we noticed that there are 4 beneficiary bank account missing, consequentely 4 beneficiary bank country missing. I'm still not sure what that means, however, it raises questions as I wonder if they are somehow corelated. 

```python
# Pulling data that only contains missing information from initial 
# dataframe

dk_dataNull = dk_data[dk_data.isnull().any(axis=1)]
dk_dataNull
```

```python
# Count of the unique payers name 
# Name of the unique payers name 

print(dk_dataNull.payer_name.nunique())
dk_dataNull.payer_name.unique()
```

```python
# Count of the unique beneficiary name 
# Name of the unique beneficiary name 

print(dk_dataNull.beneficiary_name.nunique())
dk_dataNull.beneficiary_name.unique()
```
